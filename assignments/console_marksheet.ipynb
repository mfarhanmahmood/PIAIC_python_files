```python
total = 75
def get_marks(course):
    marks = 76
    not_num = True
    loop = True
    while(loop):
        input_str = input(f"Enter {course} course marks, max({total}): ")
        
        if input_str.isdigit():
            not_num = False
            marks = int(input_str)
        else:
            not_num = True
        
        if marks > total or not_num:
            loop = True
        else:
            loop = False
    return marks
name = ""
while(len(name) <= 3):
    name = input("Enter your name: ")
    if(len(name) <= 3):
        print("Name should be greater than 3 character")

fathers_name = ""
while(len(fathers_name) <= 3):
    fathers_name = input("Enter your Father's name: ")
    if(len(fathers_name) <= 3):
        print("Father's name should be greater than 3 character")


course = "PIAIC Quarter 1"
        

marks = [get_marks("AI for Everyone"), get_marks("Python Quiz 1"), get_marks("Python Quiz 2"), get_marks("Python Quiz 3"), get_marks("Git Quiz")]

marks_sum = sum(marks)
total_percentage = round((marks_sum / (total * len(marks)))*100, 2)
marks_average = marks_sum / len(marks)

percentages = []

for val in marks:
    percentages.append(round((val / total)*100, 2))


display_message = '''\n\n\n
    ____________________________________________________________________________________
    
    \tName: {}
    \tFather's Name: {}
    \tCourse: {}
    \t
    \t------------------------------------------------------------------------
    \tCourse\t\t---\tTotal\t---\tMarks\t---\tPercentage\t
    \t------------------------------------------------------------------------
    \tAI for Everyone\t---\t75\t---\t{}\t---\t{}%\t
    \tPython Quiz 1\t---\t75\t---\t{}\t---\t{}%\t
    \tPython Quiz 2\t---\t75\t---\t{}\t---\t{}%\t
    \tPython Quiz 3\t---\t75\t---\t{}\t---\t{}%\t
    \tGit Quiz\t---\t75\t---\t{}\t---\t{}%\t
    \t------------------------------------------------------------------------\t
    \t  Total\t\t---\t375\t---\t{}\t---\t{}%\t
    \t------------------------------------------------------------------------\t
    ___________________________________________________________________________________
    
    Your average marks are: {}
'''

display_message = display_message.format(name, fathers_name, course,
                                         marks[0], percentages[0], marks[1], percentages[1], marks[2], 
                                         percentages[2], marks[3], percentages[3], marks[4], percentages[4], 
                                         marks_sum, total_percentage, marks_average)

print(display_message)
```
